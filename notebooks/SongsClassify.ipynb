```python

import nltk
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
from sinling import SinhalaTokenizer
from sinling import SinhalaStemmer

tokenizer = SinhalaTokenizer()
stemmer = SinhalaStemmer()
```

# English Text

```python
def clean_text(text):
    if not isinstance(text,str):
        return ""
    
    text = text.lower()
    
    text = re.sub(r'[^\w\s\d]','',text)
    
    words =word_tokenize(text)
    
    stop_words = set(stopwords.words('english'))
    words = [word for word in words if word not in stop_words]
    
    lemmatizer = WordNetLemmatizer()
    words = [lemmatizer.lemmatize(word) for word in words]
    
    # join the words back together
    
    text = ''.join(words)
    return text
```

```python
songs_df['Singer(s)'] = songs_df['Singer(s)'].apply(clean_text)
songs_df['Source_Domain'] =songs_df['Source_Domain'].apply(clean_text)
songs_df['Target_Domain'] =songs_df['Target_Domain'].apply(clean_text)
songs_df['Meaning'] =songs_df['Meaning'].apply(clean_text)
```

```python
def sinhala_clean_text(text):
    if not isinstance(text,str):
        return ""
    text = re.sub(r'[^\w\s\d]','',text)
    
    tokenizer = SinhalaTokenizer()
    
    text_tokenize = tokenizer.tokenize(text)
    print(text_tokenize)
    
    words = [word for word in text_tokenize if word not in stopwords_set]
    print(words)
    
    stemmer = SinhalaStemmer()
    
    stem_words = []
    for word in words:
        stemmed_result = stemmer.stem(word)
        if isinstance(stemmed_result, tuple):
            stem_words.append(stemmed_result[0]) 
        else:
            stem_words.append(stemmed_result)
    
    
    text = ''.join(stem_words)
    
    print(text)
    return text
    
    
```
